#  Потребление алкоголя студентами

Набор данных представляет собой результаты опроса 395 студентов средней школы, изучающих математику. Содержит социальную, гендерную и учебную информацию об учениках.

Атрибуты:

1.	school -  школа ('GP' - Gabriel Pereira или 'MS' - Mousinho da Silveira)
2.	sex - пол ('F' - женщина или 'M' - мужчина)
3.	age - возраст студента (от 15 до 22)
4.	address - адрес студента  ('U' - городской или 'R' - сельский)
5.	famsize - количество членов семьи ('LE3' - 3 или меньше или 'GT3' - больше трех)
6.	Pstatus - статус сожительства родителей ('T' - живут вместе или 'A' - раздельно)
7.	Medu - образование матери (0 - нет, 1 - начальное образование (4 класса), 2 - от 5 до 9 классов, 3 - среднее образование или 4 - высшее образование)
8.	Fedu - образование отца (0 - нет, 1 - начальное образование (4 класса), 2 - от 5 до 9 классов, 3 - среднее образование или 4 - высшее образование)
9.	Mjob - работа матери ('teacher', 'health', 'services', 'at_home' или 'other')
10.	Fjob - работа отца ('teacher', 'health', 'services', 'at_home' или 'other')
11.	reason - причина выбора этой школы ('home’ – близко к дому, 'reputation' – репутация школы, 'course' или  'other')
12.	guardian - опекун студента ('mother', 'father' or 'other')
13.	traveltime - время пути до школы (1 - <2 часов, 2 – от 2 до 5 часов, 3 – от 5 до 10 часов или 4 - >10 часов)
14.	studytime - количество учебы в неделю (1 - <2 часов, 2 – от 2 до 5 часов, 3 – от 5 до 10 часов или 4 - >10 часов)
15.	failures - количество прошлых неудач в классе (n если 1<=n<3, иначе 4)
16.	schoolsup – дополнительная образовательная поддержка (yes или no)
17.	famsup – семейная образовательная поддержка (yes или no)
18.	paid – дополнительные оплачиваемые уроки по математике (yes или no)
19.	activities – внеклассные занятия (yes или no)
20.	nursery – посещал ли студент детский сад (yes или no)
21.	higher – хочет ли студент получить высшее образование (yes или no)
22.	internet – есть ли доступ к интернету дома (yes или no)
23.	romantic – состоит ли студент в отношениях (yes или no)
24.	famrel – отношения в семье (от 1 – очень плохие до 5 – отличные)
25.	freetime - свободное время после школы (от 1 – очень мало до 5 – очень много)
26.	goout – прогулки с друзьями (от 1 – очень мало до 5 – очень много)
27.	Dalc – потребление алкоголя в течение рабочего дня (от 1 – очень мало до 5 – очень много)
28.	Walc - потребление алкоголя в выходные (от 1 – очень мало до 5 – очень много)
29.	health – текущее состояние здоровья (от 1 – очень плохое до 5 – очень хорошее)
30.	absences – количество прогулов (от 0 до 93)
31.	G1 – оценка по математике за первый период (от 0 до 20)
32.	G1 - оценка по математике за второй период (от 0 до 20)
33.	G3 - итоговая оценка по математике (от 0 до 20)

## Нормализация количественных признаков

Выведем количественные признаки

```python
data_numerical = data[numerical_columns]
data_numerical.describe()
```

Нормализованные количественные признаки:

```python
data_numerical = (data_numerical - data_numerical.mean(axis = 0))/data_numerical.std(axis = 0)
data_numerical.describe()
```
